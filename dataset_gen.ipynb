```python
import pandas as pd
import numpy as np

# Number of samples
n = 50000

# Set seed for reproducibility
np.random.seed(42)

# Define parent pools
flora_parents = ['A', 'B', 'C']  # A: AA, B: Aa, C: aa
fauna_parents = ['X', 'Y']       # X: short tail brown, Y: long tail black

# Sample organism types
organism_type = np.random.choice(['Flora', 'Fauna'], size=n, p=[0.5, 0.5])

# Initialize lists
parent_ids = []
genotypes = []
phenotypes = []
gametes = []

for org in organism_type:
    if org == 'Flora':
        pid = np.random.choice(flora_parents)
        if pid == 'A':
            geno = 'AA'
            gam = 'A'
        elif pid == 'B':
            geno = 'Aa'
            gam = np.random.choice(['A', 'a'])
        else:  # 'C'
            geno = 'aa'
            gam = 'a'

        parent_ids.append(pid)
        genotypes.append(geno)
        phenotypes.append('NA')
        gametes.append(gam)

    else:  # Fauna
        pid = np.random.choice(fauna_parents)
        if pid == 'X':
            phen = 'TailShort,Brown'
            gam = 'P'
        else:  # 'Y'
            phen = 'TailLong,Black'
            gam = 'Q'

        parent_ids.append(pid)
        genotypes.append('NA')
        phenotypes.append(phen)
        gametes.append(gam)

# Create DataFrame
df = pd.DataFrame({
    'OrganismType': organism_type,
    'ParentID': parent_ids,
    'Genotype': genotypes,
    'PhenotypeFeatures': phenotypes,
    'GameteType': gametes
})

# Save to CSV
file_path = '/mnt/data/gamete_dataset.csv'
df.to_csv(file_path, index=False)

# Display a preview
df.head()
```
